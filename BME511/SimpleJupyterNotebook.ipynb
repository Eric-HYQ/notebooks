```python
import numpy as np
import pylab as pl

t = np.arange(0, 1, 0.001)
x = np.sin(2 * np.pi * 10. * t)

pl.plot(t, x)
pl.xlabel('Time (s)', fontsize=14)
pl.ylabel('x(t)', fontsize=14)
```
